## 7장 군집화
### 7-1 K평균 알고리즘 이해
k-평균 알고리즘이란? \
군집 중심점이라는 특정한 임의의 지점을 선택해 해당 중심에 가장 가까운 포인트들을 선택하는 군집화 기법. \
선택된 포인트의 평균 지점으로 이동하고 이동된 중심점에서 다시 가까운 포인트를 선택, 다시 중심점을 평균 지점으로 이동하는 프로세스를 반복.  
모든 데이터 포인트에서 더 이상 중심점이 없을 경우 반복을 멈추고 해당 중심점에 해당하는 데이터 포인트들을 군집화하는 기법. \
 
장점 : 일반적인 군집화에서 가장 많이 활용, 알고리즘이 쉽고 간단. 
단점 : 거리기반 알고리즘으로 속성의 개수가 매우 많을 경우 군집화 정확도가 떨어짐. 반복 횟수가 많을 경우 수행 시간이 느려짐. 몇개의 군집을 선택할지 가이드가 어려움.


### 7-2 군집 평가
1. 실루엣 분석 : 각 군집간 거리가 얼마나 효율적으로 분리돼 있는지를 나타냄. 효율적으로 분리가 됐다는 것은 다른 군집과의 거리는 떨어져있고 동일 군집간 데이터는 가깝게 뭉쳐있음을 의미. \
    실루엣 계수 : 
   단점 : 각 데이터별로 다른 데이터와의 거리를 반복적으로 계산해야 하므로 데이터 양이 늘어나면 수행시간이 크게 증가. 



### 7-3 평균 이동
k-평균은 중심이 소속된 데이터의 평균 거리 중심으로 이동하지만, 평균 이동은 중심을 데이터가 모여 있는 밀도가 가장 높은 곳으로 이동. \
확률밀도 함수를 사용하여, 가장 집중적으로 데이터가 모여있어 확률 밀도 함수가 피크인 점을 군집 중심점으로 선정하며 일반적으로 주어진 모델의 확률 밀도 함수를 찾기 위해 KDE 사용. \ 
평균 이동 군집화 : 특정 데이터를 반경 내의 데이터 분포 확률 밀도가 가장 높은 곳으로 이동하기 위해 주변 데이터와의 거리 값을 KDE 함수 값으로 입력한 뒤 그 반환 값을 현재 위치에서 업데이트하면서 이동하는 방식을 취함. \

 \ 
KDE는 다음과 같은 커널 함수식으로 표현됨. \ 
 \
h 값이 작으면 좁고 뾰족한 KDE, 매우 크면 평활한 KDE 값을 가짐. 대역 폭이 클수록 군집 중심점이 적음. 


### 7-4 GMM
GMM : 군집화를 적용하고자 하는 데이터가 여러 개의 가우시안 분포를 가진 데이터 집합들이 섞여서 생성된 것이라는 가정하에, 군집화를 수행하는 방식. \
GMM 과 K-평균의 비교 : GMM은 k-평균보다 유연하게 다양한 데이터 세트에 적용 가능하지만, 군집화를 위한 수행 시간이 오래 걸림. 

### 7-5 DBSCAN
DBSCAN : 밀도 기반 군집화의 대표적인 알고리즘. 복잡한 기하학적 분포도를 가진 데이터 세트에 대해서도 군집화를 효과적으로 수행. 입실론 주변 영역의 최소 대아터 개수를 포함하는 밀도 기준을 충족시키는 핵심 포인트를 연결하면서 군집화를 구성하는 방식 \
 \
주요 파라미터
1) epsilon : 개별 데이터를 중심으로 입실론 반경을 가지는 원형의 영역
2) min points : 개별 데이터의 입실론 주변 영역에 포함되는 타 데이터의 개수

데이터 포인트
1) 핵심 포인트 : 주변 영역 내 최소 데이터 개수 이상의 타 데이터를 가지고 있을 경우 해당 데이터를 핵심 포인트라 함. \
2) 이웃 포인트 : 주변 영역 내에 위치한 타 데이터
3) 경계 포인트 : 주변 영역 내에 최소 데이터 개수 이상의 이웃 포인트를 가지고 있진 않지만, 핵심 포인트를 이웃 포인트로 가짐 \
4) 잡음 포인트 : 최소 데이터 개수 이상의 이웃 포인트를 가직 있지 않으며, 핵심 포인트도 이웃 포인트로 가지고 있지 않는 데이터. \
