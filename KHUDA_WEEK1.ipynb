### 넘파이의 ndarray


```python
import numpy as np
```

```python
# np.array() : ndarray로 변환을 원하는 객체를 인자로 입력하면, ndarray를 반환

array1 = np.array([1,2,3])
print(array1)
```

```python
# ndarray.shape : nadarray의 차완과 크기를 튜플 형태로 반환
print(array1.shape)
```

```python
array2 = np.array([[1,2,3], [4,5,6]])
print(array2)
print(array2.shape)
```

```python
# np.astype() : 괄호 안에 인자로 원하는 타입을 지정
array_int = np.array([1,2,3])
array_float = array_int.astype('float64')
print(array_float, array_float.dtype)
```

### ndarray를 편리하게 생성하기 - arange, zeros, ones

```python
sequence_array = np.arange(10)
print(sequence_array) # 0부터 함수 인자 값 -1까지 값을 순차적으로 ndarray의 데이터 값으로 변환
```

```python
#np.zeros() : 인자로 튜플 형태의 shape 값을 입력하면, 모든 값을 0으로 할당 
#np.ones() : zeros와 마찬가지로 모든 값을 1로 할당
zero_array = np.zeros((3,2), dtype='int32')
print(zero_array)
print(zero_array.dtype, zero_array.shape)

one_array = np.ones((3,2))
print(one_array)
print(one_array.dtype, one_array.shape)
```

### ndarray의 차원과 크기를 변경하는 reshape

(10,) 데이터를 (4,3) 형태로 변경은 불가. reshape의 인자를 -1로 적용여 많이 씀.
reshape(-1,1) : 원본 ndarray가 어떤 형태 든, 2차원이고 여러개의 로우를 가지되 반드시 1개의 컬럼을 가진 ndarray로 반환

### 넘파이의 ndarray 인덱싱
* 단일 값 추출 : 원하는 위치의 인덱스 값을 지정하면 해당 위치의 데이터 반환 - array1[2]
* 슬라이싱 : 연속된 인덱스상의 ndarray를 추출하는 방식 - array1[0:3]
* 팬시 인덱싱 : 일정한 인덱싱 집합을 리스트 또는 ndarray 형태로 지정해 해당 위치의 데이터를 ndarray로 반환 - array1[[0,1], 0:2]
* 불린 인덱싱 : 특정 조건에 해당하는지 TRUE/FALSE 값을 인덱싱 집합을 기반으로 True에 해당하는 인덱스 위치에 있는 데이터의 ndarray를 반환. array1[array1>5]


sort() : 원 행렬은 그대로 유지한 채 원행렬의 정렬된 행렬을 반환.
argsort() : 정렬 행렬의 원본 행렬의 인덱스를 ndarray형으로 반환

```python
# 멧플롯립 (Matplotlib)
import matplotlib.pyplot as plt

# x좌표 입력값 [1, 2, 3], y좌표 입력값 [2, 4, 6]으로 그래프를 그리기. 
plt.plot([1, 2, 3], [2, 4, 6]) 

plt.title("Hello plot") 

# 선 그래프를 출력
plt.show()
```

figure : 그림판의 크기 조절, 플롯을 최종적으로 화면에 나타내는 역할 \
axes : 실제 그림을 그리는 메서드들을 가짐. X,Y축, 타이틀, 범례 등의 속성을 설정하는데 사용


```python
plt.figure(figsize=(10, 4)) 

plt.plot([1, 2, 3], [2, 4, 6]) 
plt.title("Hello plot") 
plt.show()
```

```python
# 여러개의 plot을 갖는 subplot 설정
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(6, 3))
```

```python
x_value_01 = np.arange(1, 10)
y_value_01 = 2*x_value_01

# 마커를 포함한 빨간선 대쉬 선 그래프를 그리고, label은 'red line'으로 설정  
plt.plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
# X값에 따른 Y값을 나타내는 초록색 막대 그래프를 그리고 label은 'bar plot'으로 설정
plt.bar(x_value_01, y_value_01, color='green', label='bar plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
# 두개의 그래프에서 label설정된 문자열 값을 범례로 표시
plt.legend()

plt.title('Hello plot')

plt.show()
```

축의 명칭 설정 : xlabel, ylabel 사용

```python
import numpy as np

x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02

# figsize는 (6,6)인 Figure와 2x2개의 Axes를 가지는 subplots을 반환.
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6, 6))

# 2x2 Axes객체들을 각각 이용하여 선 그래프와 막대그래프를 그림. 
ax[0][0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax[0][1].bar(x_value_02, y_value_02, color='green', label='green bar')
ax[1][0].plot(x_value_01, y_value_01, color='green', marker='o', linestyle='dashed', label='green line')
ax[1][1].bar(x_value_02, y_value_02, color='red', label='red bar')

# 개별 Axes객체 각각에 X축명을 설정. 
ax[0][0].set_xlabel('ax[0][0] x axis')
ax[0][1].set_xlabel('ax[0][1] x axis')
ax[1][0].set_xlabel('ax[1][0] x axis')
ax[1][1].set_xlabel('ax[1][1] x axis')

# 개별 Axes객체 각각에 범례를 설정 
ax[0][0].legend()
ax[0][1].legend() 
ax[1][0].legend()
ax[1][1].legend() 

plt.show()
```
