### 3-1 정확도

정확도 : 실제 데이터에서 예측 대이터가 얼마나 같은지를 판단하는 지표 \
 \
불균형한 레이블 데이터 세트에서는 성능 수치로 사용 불가 

## 3-2 오차 행렬
오차 행렬 : 학습된 분류 모댈이 예측을 수행하면서 얼마나 confused 되는지도 보여줌. \
이진 분류의 예측 오류가 얼마인지와 더불어 어떠한 유형의 예측 오류가 발생하고 있는지를 함께 나타내는 지표 \
 \
TN : negative -> negative로 예측 \
FP : negative -> positive로 예측 \
TN : positive -> negative로 예측 \
TP : positive -> positive로 예측 \
정확도 = (TN + TP)/(TN + FP + FN + TP) 





### 3-3 정밀도와 재현율
정밀도 = TP/(FP + TP) \
재현율 = TP/(FN + TP) \
재현율이 중요한 경우(보통의 경우) : positive를 negative로 판단 시 큰 영향을 끼치는 경우. 암 판단 모델이, 금융 사기 적발 모델 \ 
정밀도가 중요한 경우 : negative 인 데이터를 positive로 판단 시 큰 영향을 끼치는 경우, 스팸 메일 판단 모델 \
정밀도/재현율 트레이드오프 : 한 쪽을 강제로 높이면 다른 하나의 수치는 떨어지는 현상 \
 \
사이킷런의 API = precision_recall_curve() \



### 3-4 F1스코어
F1 스코어 : 정밀도와 재현율을 결합한 지표 \
 



### 3-5 ROC 곡선과 AUC
민감도 TPR : 실제값 positive가 정확히 예측 되어야 하는 수준. TN/(FP+TN) \
특이성 TNR : 실제값 negative가 정확히 예측 되어야 하는 수준. FP/(FP+TN) \
FPR = FP / (FP + TN) = 1 - TNR = 1- 특이성 \
AUC : ROC곡선의 밑 면적. 1에 가까울 수록 좋은 수치.



# 4장 
### 4-1 분류의 개요
분류 : 지도학습의 대표적 유형, 주어진 데이터의 피처와 레이블 값을 머신러닝 알고리즘으로 학습해 모델을 생성하고, 생성된 모델에 새로운 데이터 값이 주어졌을 때 미지의 레이블 값을 예측하는 것. \




### 4-2 결정 트리
데이터에 있는 규칙을 학습을 통해 자동으로 찾아내 트리 기반의 분류 규칙을 만드는 것. \
많은 규칙이 있으면 과적합으로 이어지기 쉽다. 깊이가 깊어질 수록, 결정 트리의 예측 성능이 저하될 가능성이 있음 \ 
 \
정보의 균일도를 측정하는 대표적인 방법 : 정보 이득 지수, 지니 계수 \
정보 이득 : 엔트로피(주어진 데이터 집합의 혼잡도) 개념을 기반으로 서로 다른 값이 섞여있으면 엔트로피 값이 높고, 같은 값이 섞이면 엔트로피 값이 낮음. 정보 이득 지수는 1 - 엔트로피 지수. 정보 이득이 높은 속성을 기준으로 분할 \
지니 계수 : 0이 가장평등, 1로 갈수록 불평동. 지니 계수가 낮을 수록 데이터 균일도가 높다고 해석해 지니 계수가 낮은 속성을 기준으로 분할.  
결정 트리 장점 : 쉽고 직관적임. 피처 스케일링이나 정규화 등의 사전 가공 영향도가 크지 않음. \
결정 트리 단점 : 과적합으로 알고리즘 성능이 떨어짐. 트리의 크기를 사전에 제한하는 튜닝 필요. 



### 4-3 앙상블 학습
비정형 데이터의 분류는 딥러닝이 뛰어난 성능을 보임. 대부분의 정형 데이터 분류 시에는 앙상블이 뛰어난 성능을 보임. \ 
앙상블 알고리즘의 예시 : 랜덤포레스트, 그라디언트 부스팅 알고리즘. XGBoost, LightGBM, Stacking \
앙상블 학습의 유형 : 보팅, 배깅, 부스팅으로 크게 나눌 수 있음. \
 \
Bootstrapping : 개별 classifier에 데이터를 샘플링 해서 추출하는 방식 \  
보팅 유형 - 하드 보팅, 소프트 보팅 \ 
하드 보팅 : 예측한 결과값 중 다수의 분류기가 결정한 예측값을 최종 보팅 결과값으로 선정. \  
소프트 보팅 : 분류기들의 레이블 값 결정 확률을 모두 더하여 평균을 내 확률이 가장 높은 레이블 값을 최종 보팅 결과값으로 선정. \
 

    


### 4-4 랜덤 포레스트
배깅 : 같은 알고리즘으로 여러 개의 분류기를 만들어서 보팅으로 결정하는 알고리즘. 대표적인 예시는 랜덤 포레스트 \
랜덤 포레스트 : 여러 개의 결정 트리 분류기가 전체 데이터에서 배깅 방식으로 각자의 데이터를 샘플링해 개별적으로 학습을 수행한 후 최종적으로 모든 분류기가 보팅을 통해 예측 결정을 함. \  
 \
부트스트래핑 : 여러 개의 데이터 세트를 중첩되게 분리하는 것 \ 




### 4-5 GBM 
부스팅 알고리즘 : 여러 개의 약한 학습기를 순차적으로 학습-예측 하면서 잘 못 예측한 데이터에 가중치를 부여해 오류를 개선해 나가면서 학습하는 방식. 대표적 예시는 AdaBoost \
GBM은 분류 및 회귀도 가능.
